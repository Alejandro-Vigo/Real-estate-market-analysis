# Project Design

This case simulates that we are a real estate company that invests in large cities by purchasing properties to later rent them out as tourist apartments.

Management has decided to invest in Madrid and has tasked us with analyzing the data published by industry leader Airbnb to try to find the types of properties that have the greatest commercial potential for tourist rentals.

The main deliverable expected is the type (or types) of properties that the valuation team should look for among the opportunities available in the city and the main neighborhoods or geographical areas to focus on.

Although this specific case focuses on tourist rentals, the same type of approach can be used in cases that have a high “location” component:

* opening and closing of stores
* reduction of installed capacity
* expansion of franchises
* etc.

## Objective

Identify the property profile(s) that maximize commercial potential in the tourist rental market and the main areas where to look for them.

## Levers

* **Rental price**: the more you can charge per night, the higher the return
* **Occupancy**: in general, the more days per year a property can be rented out, the higher the return
* **Property price**: the cheaper the property can be purchased, the higher the return

## KPIs

* We will measure occupancy as the number of days per year that the property can be rented
* We will measure the rental price as the price per night in euros according to Airbnb
* We will measure the price of a property as the multiplication between the number of square meters and the average price per square meter in its area, and we will apply a 25% discount on the official price due to the negotiating power of our purchasing team.

## Entities and data

The entities relevant to our objective and for which we have data available are:

* Properties
* Owners
* Districts

## Seed Questions

About rental prices:

* What is the average price? What is the price range? What about by district? What about by neighborhood?
* What is the ranking of districts and neighborhoods by average rental price?
* What factors (apart from location) determine rental prices?
* What is the relationship between the size of the property and the price at which it can be rented?
* How does competition (number of properties available per neighborhood) influence the rental price?
* How do prices vary by type of rental (entire apartment, private room, shared room)?


About occupancy:

* What is the average occupancy? And by district? And by neighborhood?
* How likely is each occupancy level in each district?
* What is the ranking of districts and neighborhoods by occupancy?
* What factors (apart from location) determine occupancy?
* What is the relationship between the size of the property and its occupancy rate?
* How does competition (number of properties available per neighborhood) influence occupancy?

Regarding the purchase price:

* What is the ranking of price per square meter by district?
* What is the ranking of property price (square meters * average size) by district?
* What is the relationship between property price and rental price by district?
* What is the relationship between property price and occupancy by district?
